# Simulaci√≥n del problema de Monty Hall

```python
import random

#se puede ajustar estas dos variables desde aca para obtener distintas simulaciones
cantPuertas = 3
cantIntentos = 1000

aciertosAlCambiar = 0
aciertosAlQuedarse = 0
aciertosAlElegirAlAzar = 0

acumAciertosAlCambiar = []
acumAciertosAlQuedarse = []
acumAciertosAlElegirAlAzar = []

for i in range(cantIntentos):
    
    #definimos las puertas
    numPuertaSeleccionada = random.randint(1, cantPuertas)
    numPuertaConPremio = random.randint(1, cantPuertas)
    numPuertaAbierta = random.choice([num for num in range(1,cantPuertas+1) if num not in [numPuertaSeleccionada, numPuertaConPremio]])

    #calcular aciertos al no cambiar de eleccion
    if numPuertaSeleccionada == numPuertaConPremio:
        aciertosAlQuedarse = aciertosAlQuedarse+1
    
    #calcular aciertos al cambiar de puerta
    numPuertaSeleccionada = random.choice([num for num in range(1,cantPuertas+1) if num not in [numPuertaSeleccionada, numPuertaAbierta]])
    if numPuertaSeleccionada == numPuertaConPremio:
        aciertosAlCambiar = aciertosAlCambiar + 1
    
    #calcular aciertos al elegir al azar entre todas las puertas cerradas
    if random.choice([num for num in range(1,cantPuertas+1) if num != numPuertaAbierta]) == numPuertaConPremio:
        aciertosAlElegirAlAzar = aciertosAlElegirAlAzar + 1

    acumAciertosAlQuedarse.append(aciertosAlQuedarse)
    acumAciertosAlCambiar.append(aciertosAlCambiar)
    acumAciertosAlElegirAlAzar.append(aciertosAlElegirAlAzar)
    
print(aciertosAlCambiar, aciertosAlQuedarse, aciertosAlElegirAlAzar)


    
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(acumAciertosAlQuedarse, label='Sin cambiar')
plt.plot(acumAciertosAlCambiar, label='Cambiando')
plt.plot(acumAciertosAlElegirAlAzar, label='Decision al azar')
plt.xlabel('Itentos')
plt.ylabel('Aciertos')
plt.legend()
plt.show()
```
